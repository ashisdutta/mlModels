```python
import pandas as pd
df = pd.read_csv("E:/6th sem lab/ml lab/lab3p1CSV.csv")
df
```

```python
df.drop(['Name'], axis='columns', inplace =True)
df
```

```python
dummies = pd.get_dummies(df.Gender)
dummies
```

```python
df=pd.concat([df, dummies], axis='columns')
df
```

```python
df.drop('Gender', axis = 'columns', inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['Age', 'Chol', 'Female', 'Male']], df.Target,test_size = 0.2)
```
